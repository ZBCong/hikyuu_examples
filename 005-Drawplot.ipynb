四个创建组合窗口的快捷函数,返回相应的 axes 对象tuple，可在绘制K线、指标、系统信号的 plot 方法中指定对应的 axes：
- create_one_axes_figure
- create_two_axes_figure
- create_three_axes_figure
- create_four_axes_figure

```python
s = sm['sh000001']
k = s.getKData(Query(-200))
ax1,ax2 = create_two_axes_figure()
k.plot(axes=ax1)

ma = EMA(CLOSE(k))
ma.plot(axes=ax1, legend_on=True)

v = VIGOR(KDATA(k))
v.plot(axes=ax2, legend_on=True)
```

绘制MACD：
- ax_draw_macd
- ax_draw_macd2

```python
ax1,ax2, ax3 = create_three_axes_figure()

k.plot(axes=ax1)
ma.plot(axes=ax1, legend_on=True)

v.plot(axes=ax2, legend_on=True)

ax_draw_macd(axes=ax3, kdata=k)
```
